### Exercise 1: Line Plot Customization

```python
# TASK: Create a line plot with x values ranging from 0 to 10 and y values as the square of x.
# Customize the plot by adding a title, labels for both axes, and a grid.
```

```python
import matplotlib.pyplot as plt

# x values from 0 to 10
x = list(range(0, 11))

# y values as square of x
y = [i ** 2 for i in x]

# Create the plot
plt.plot(x, y, marker='o', color='b', linestyle='--')

# Customize the plot
plt.title("Line Plot: x vs xÂ²")
plt.xlabel("x values")
plt.ylabel("x squared")
plt.grid(True)

# Show the plot
plt.show()
```
